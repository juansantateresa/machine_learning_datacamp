```python
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('seeds.csv')
```

```python
variety_numbers = df['1'].to_list()
```

```python
variety_numbers
```

```python
samples = df.drop(columns = ['1'])
```

```python
samples
```

```python
model = TSNE(learning_rate=200)
```

Let's fit the model to samples 


```python
tsne_features = model.fit_transform(samples)
```

0th feature & 1st feature

```python
xs = tsne_features[:,0]
```

```python
ys = tsne_features[:,1]
```

```python
plt.scatter(xs, ys, c=variety_numbers)
```
