```python
df_2 = pd.read_csv('wikipedia-vectors.csv')
```

```python
df_2_t= df_2.T
```

```python
df_2_t
```

```python
model = NMF(n_components=6)
```

```python
model.fit(df_2_t)
```

```python
nmf_features = model.transform(df_2_t)
```

```python
print(nmf_features.round(2))
```
